## Grover's algorithm on a 4-qubit space using the V-Oracle model

Search for the state |1111>

Import the required packages

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from qiskit import Aer, assemble
from qiskit.visualization import plot_histogram
```

Define the registers

```python
qreg_q = QuantumRegister(6, 'q')
creg_c = ClassicalRegister(4, 'c')
circuit = QuantumCircuit(qreg_q, creg_c)
```

Initialise the states

```python
circuit.h(qreg_q[0])
circuit.h(qreg_q[1])
circuit.h(qreg_q[3])
circuit.h(qreg_q[5])
```

Quantum barrier

```python
circuit.barrier(qreg_q[0], qreg_q[1], qreg_q[2], qreg_q[3], qreg_q[4], qreg_q[5])
```

Iteration 1

```python
# Oracle iteration 1
circuit.ccx(qreg_q[0], qreg_q[1], qreg_q[2])
circuit.ccx(qreg_q[2], qreg_q[3], qreg_q[4])
circuit.cz(qreg_q[4], qreg_q[5])
circuit.ccx(qreg_q[2], qreg_q[3], qreg_q[4])
circuit.ccx(qreg_q[0], qreg_q[1], qreg_q[2])
# barrier
circuit.barrier(qreg_q[0], qreg_q[1], qreg_q[2], qreg_q[3], qreg_q[4], qreg_q[5])
# diffuser iteration 1
circuit.h(qreg_q[0])
circuit.h(qreg_q[1])
circuit.h(qreg_q[3])
circuit.h(qreg_q[5])
circuit.x(qreg_q[0])
circuit.x(qreg_q[1])
circuit.x(qreg_q[3])
circuit.x(qreg_q[5])
circuit.ccx(qreg_q[0], qreg_q[1], qreg_q[2])
circuit.ccx(qreg_q[2], qreg_q[3], qreg_q[4])
circuit.cz(qreg_q[4], qreg_q[5])
circuit.ccx(qreg_q[2], qreg_q[3], qreg_q[4])
circuit.ccx(qreg_q[0], qreg_q[1], qreg_q[2])
circuit.x(qreg_q[0])
circuit.x(qreg_q[1])
circuit.x(qreg_q[3])
circuit.x(qreg_q[5])
circuit.h(qreg_q[0])
circuit.h(qreg_q[1])
circuit.h(qreg_q[3])
circuit.h(qreg_q[5])
# barrier
circuit.barrier(qreg_q[0], qreg_q[1], qreg_q[2], qreg_q[3], qreg_q[4], qreg_q[5])
circuit.barrier(qreg_q[0], qreg_q[1], qreg_q[2], qreg_q[3], qreg_q[4], qreg_q[5])
```

Iteration 2

```python
# Oracle iteration 2
circuit.ccx(qreg_q[0], qreg_q[1], qreg_q[2])
circuit.ccx(qreg_q[2], qreg_q[3], qreg_q[4])
circuit.cz(qreg_q[4], qreg_q[5])
circuit.ccx(qreg_q[2], qreg_q[3], qreg_q[4])
circuit.ccx(qreg_q[0], qreg_q[1], qreg_q[2])
# barrier
circuit.barrier(qreg_q[0], qreg_q[1], qreg_q[2], qreg_q[3], qreg_q[4], qreg_q[5])
# diffuser iteration 2
circuit.h(qreg_q[0])
circuit.h(qreg_q[1])
circuit.h(qreg_q[3])
circuit.h(qreg_q[5])
circuit.x(qreg_q[0])
circuit.x(qreg_q[1])
circuit.x(qreg_q[3])
circuit.x(qreg_q[5])
circuit.ccx(qreg_q[0], qreg_q[1], qreg_q[2])
circuit.ccx(qreg_q[2], qreg_q[3], qreg_q[4])
circuit.cz(qreg_q[4], qreg_q[5])
circuit.ccx(qreg_q[2], qreg_q[3], qreg_q[4])
circuit.ccx(qreg_q[0], qreg_q[1], qreg_q[2])
circuit.x(qreg_q[0])
circuit.x(qreg_q[1])
circuit.x(qreg_q[3])
circuit.x(qreg_q[5])
circuit.h(qreg_q[0])
circuit.h(qreg_q[1])
circuit.h(qreg_q[3])
circuit.h(qreg_q[5])
# barrier
circuit.barrier(qreg_q[0], qreg_q[1], qreg_q[2], qreg_q[3], qreg_q[4], qreg_q[5])
circuit.barrier(qreg_q[0], qreg_q[1], qreg_q[2], qreg_q[3], qreg_q[4], qreg_q[5])
```

Iteration 3

```python
# Oracle iteration 3
circuit.ccx(qreg_q[0], qreg_q[1], qreg_q[2])
circuit.ccx(qreg_q[2], qreg_q[3], qreg_q[4])
circuit.cz(qreg_q[4], qreg_q[5])
circuit.ccx(qreg_q[2], qreg_q[3], qreg_q[4])
circuit.ccx(qreg_q[0], qreg_q[1], qreg_q[2])
# barrier
circuit.barrier(qreg_q[0], qreg_q[1], qreg_q[2], qreg_q[3], qreg_q[4], qreg_q[5])
# diffuser iteration 3
circuit.h(qreg_q[0])
circuit.h(qreg_q[1])
circuit.h(qreg_q[3])
circuit.h(qreg_q[5])
circuit.x(qreg_q[0])
circuit.x(qreg_q[1])
circuit.x(qreg_q[3])
circuit.x(qreg_q[5])
circuit.ccx(qreg_q[0], qreg_q[1], qreg_q[2])
circuit.ccx(qreg_q[2], qreg_q[3], qreg_q[4])
circuit.cz(qreg_q[4], qreg_q[5])
circuit.ccx(qreg_q[2], qreg_q[3], qreg_q[4])
circuit.ccx(qreg_q[0], qreg_q[1], qreg_q[2])
circuit.x(qreg_q[0])
circuit.x(qreg_q[1])
circuit.x(qreg_q[3])
circuit.x(qreg_q[5])
circuit.h(qreg_q[0])
circuit.h(qreg_q[1])
circuit.h(qreg_q[3])
circuit.h(qreg_q[5])
# barrier
circuit.barrier(qreg_q[0], qreg_q[1], qreg_q[2], qreg_q[3], qreg_q[4], qreg_q[5])
circuit.barrier(qreg_q[0], qreg_q[1], qreg_q[2], qreg_q[3], qreg_q[4], qreg_q[5])
```

 Measurements

```python
circuit.measure(qreg_q[0], creg_c[0])
circuit.measure(qreg_q[1], creg_c[1])
circuit.measure(qreg_q[3], creg_c[2])
circuit.measure(qreg_q[5], creg_c[3])
```

Simulation of the circuit

```python
aer_sim = Aer.get_backend('aer_simulator')
qobj = assemble(circuit)
result = aer_sim.run(qobj).result()
counts = result.get_counts()
plot_histogram(counts)
```

Visualization of the circuit
